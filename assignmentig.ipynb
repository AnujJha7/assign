```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_meal = pd.read_csv(r"C:\Users\Anuj Navin\Downloads\meal_info.csv")

df_center = pd.read_csv(r"C:\Users\Anuj Navin\Downloads\fulfilment_center_info.csv")   

df_food= pd.read_csv(r"C:\Users\Anuj Navin\Downloads\train.csv")

df = pd.merge(df_center, df_food, how ='outer', on= 'center_id') 

df = pd.merge(df, df_meal, how ='outer', on= 'meal_id') 

##bargraph that we need here
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
print(table)
 
    
table.columns=['A']
dataframe = table.reset_index(level = 'category')    
a= dataframe['category']
b=dataframe['A']
plt.bar(a,b)
plt.xticks(rotation=70)
plt.xlabel('Category')
plt.ylabel('Orders')

df['revenue']=df['num_orders'] * df['checkout_price']

df['months'] =  df['week']//4

print(df)
```

```python
##formonths
table1 = pd.pivot_table(data=df,index= 'months', values = 'revenue' ,aggfunc = np.sum)
table1.columns=['B']
tableframe1 = table1.reset_index(level='months')
m=tableframe1['months']
n=tableframe1['B']
plt.subplot(1,2,1)
plt.plot(m,n)
plt.ylabel('Revenue')
plt.xlabel('Months')

##forweeks

table2 = pd.pivot_table(data=df,index='week',values ='revenue', aggfunc =np.sum)
table2.columns=['Z']
dataframe2=table2.reset_index(level='week')
p=dataframe2['week']
q=dataframe2['Z']
plt.subplot(1,2,2)
plt.plot(p,q)
plt.xlabel('weeks')
plt.ylabel('revenue')
plt.rcParams["figure.figsize"] = (20,10)

 
```
